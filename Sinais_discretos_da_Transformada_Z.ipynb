```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotar_sinal_discreto(sinal: np.array, range:list=list()):
    if len(range) > 0:
      plt.stem(range, sinal, use_line_collection=True)
    else:
      plt.stem(sinal, use_line_collection=True)
    plt.title('Sinal no tempo discreto')
    plt.xlabel('Tempo discreto')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()
    plt.tight_layout()
```

## Gráficos referentes a cada resposta ao impulso da função de transferência do diagrama 3 da Lista da Unidade III

```python
n = np.linspace(-15, 15, 80)

h = ((3.58333 * (-0.667**n)) - (1.33333 * (0.33**n))) * np.heaviside(n, 1)

plotar_sinal_discreto(h, n)
```

```python
n = np.linspace(-np.pi, np.pi, 100)

h = ((3.58333) * (-0.667 ** n) * (-np.heaviside(-n-1, 1))) - ((1.33333) * (0.33**n) * np.heaviside(n, 1))

plotar_sinal_discreto(h, n)
```
